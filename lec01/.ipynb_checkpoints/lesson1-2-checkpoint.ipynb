```python
from collections import Counter

# score_checker.py参照
# 文字ごとのスコアテーブル（a〜z）
SCORES = [1, 3, 2, 2, 1, 3, 3, 1, 1, 4, 4, 2, 2,
          1, 1, 3, 4, 1, 1, 1, 2, 3, 3, 4, 3, 4]

def calculate_score(word): # 単語のスコアを計算
    score = 0
    for character in list(word):
        score += SCORES[ord(character) - ord('a')]
    return score


# 単語が入力文字から作れるか判定
def can_build_from_input(word, input_counter): # word:dictionary上の単語
    word_counter = Counter(word)
    for char, count in word_counter.items():
        if input_counter[char] < count:
            return False
    return True

# 辞書の単語をすべて調べて、最大のスコアを持つアナグラムを返す
def find_best_word(input_word, dictionary):
    input_counter = Counter(input_word)
#    max_score = 0
#    best_word = ""
    for word_with_score in dictionary: # ([30][act], [5][cat], ...)
        if can_build_from_input(word_with_score[1], input_counter) == True:
            return word_with_score[1]
        #else :
        #    continue
    return ""
    
#        if can_build_from_input(word, input_counter):
#            score = calculate_score(word)
#            if score > max_score:
#                max_score = score
#                best_word = word
#    return best_word




def dictionary_score(dictionary) :
    new_dictionary = []
    for word in dictionary :
        new_dictionary.append([calculate_score(word), word])
    new_dictionary = sorted(new_dictionary, reverse=True)
    return new_dictionary
        


### メイン処理 ###

# 辞書を読み込み
with open("words.txt") as f:
    dictionary = [line.strip() for line in f.readlines()]

# 入力を読み込み
with open("large.txt") as f:
    inputs = [line.strip() for line in f.readlines()]

new_dictionary = dictionary_score(dictionary)

# 出力ファイルに最大スコア単語を書き出し
with open("large_answer.txt", "w") as f:
    for input_word in inputs:
        best_word = find_best_word(input_word, new_dictionary)
        f.write(best_word + "\n")
```
